```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Dataset
X = np.array([1,1.5, 2, 3, 4])
Y = np.array([20,35 ,40, 60, 80])

# Step 2: Initializing Parameters
theta_0 = 0
theta_1 = 0

# Step 3: Set learning rate and iterations
alpha = 0.001
iterations = 100

# For tracking cost
costs = []

# Step 4: Gradient Descent
for i in range(iterations):
    Y_pred = theta_0 + theta_1 * X  # Predictions  theta_0 as c , theta_1 as m in y=mx+c
    error = Y_pred - Y              # Error
    cost = (1 / (2 * len(X))) * np.sum(error ** 2)  # Cost = (1/2m)*((theta_0+theta_1*x1-y1)^2 +  (theta_0+theta1*x2-y2)^2 +....  )
    costs.append(cost)

    # Gradients (by partial differentiations)
    d_theta_0 = (1 / len(X)) * np.sum(error)
    d_theta_1 = (1 / len(X)) * np.sum(error * X)

    # Update parameters
    theta_0 -= alpha * d_theta_0
    theta_1 -= alpha * d_theta_1

    # Printing every 10 iterations
    if (i + 1) % 10 == 0:
        print(f"Iteration {i+1}: θ0 = {theta_0:.4f}, θ1 = {theta_1:.4f}, Cost = {cost:.4f}")


```

```python
# Step 5: Plotting Cost over Iterations
plt.plot(range(1, iterations + 1), costs, marker='o')
plt.title('Cost Function vs Iterations')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.grid(True)
plt.show()

```

```python


# Step 6: Plotting Regression Line
plt.scatter(X, Y, color='blue', label='Actual')

plt.plot(X, theta_0 + theta_1 * X, color='red', label='Predicted Line')
plt.title('Linear Regression Fit')
plt.xlabel('X (input)')
plt.ylabel('Y (output)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Dataset with 2 features
X = np.array([
    [1, 2],
    [1.5, 1],
    [2, 3],
    [3, 2],
    [4, 4]
])  # shape (5, 2)

Y = np.array([30, 25, 45, 50, 70])  # shape (5,)

# Step 2: Add bias (x0 = 1)
m = len(Y)
X_b = np.c_[np.ones((m, 1)), X]  # shape (5, 3)

# Step 3: Initialize thetas
theta = np.zeros(X_b.shape[1])  # [theta_0, theta_1, theta_2]

# Step 4: Set learning rate and iterations
alpha = 0.01
iterations = 100
costs = []

# Step 5: Gradient Descent
for i in range(iterations):
    Y_pred = X_b.dot(theta)
    error = Y_pred - Y
    cost = (1 / (2 * m)) * np.sum(error ** 2)
    costs.append(cost)

    gradients = (1 / m) * X_b.T.dot(error)
    theta -= alpha * gradients

    if (i + 1) % 10 == 0:
        print(f"Iteration {i+1}: theta = {theta}, Cost = {cost:.4f}")

# Step 6: Plot Cost vs Iterations
plt.plot(range(1, iterations + 1), costs, marker='o')
plt.title('Cost Function vs Iterations (Multivariate)')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.grid(True)
plt.show()

# Step 7: Predict new data
new_data = np.array([
    [1, 2],
    [3, 3]
])
new_data_b = np.c_[np.ones((new_data.shape[0], 1)), new_data]
predictions = new_data_b.dot(theta)
print("Predictions on new data:", predictions)
```
